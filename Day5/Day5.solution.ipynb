## Part 1

```python
import numpy as np
import copy
```

```python
# Ingest
with open('Day5.input', 'r') as fin:
    contents = fin.readlines()

rules = []
for idx, x in enumerate(contents):
    if x != '\n':
        rule = [int(y) for y in x.strip().split('|')]
        rules.append(rule)
    else:
        break  
rules = np.array(rules)
print(f'{len(rules)} rules')

arrays = []
for line in contents[idx+1:]:
    arrays.append([int(x) for x in line.strip().split(',')])
print(f'{len(arrays)} updates')
```

```python
def test_rule(rule, sequence):
    if not (rule[0] in sequence) or not (rule[1] in sequence):
        # Pass
        return True

    a = np.where(sequence == rule[0])
    b = np.where(sequence == rule[1])
    
    if np.max(a) < np.min(b):
        # Pass
        return True

    # Fail
    return False
```

```python
# Sort arrays. Did not understand the part 1 assignment
orig = copy.deepcopy(arrays)

array_idx = 0
while(array_idx < len(arrays)):
    print(f'Sequence: {arrays[array_idx]}')
    correct = False
    while(not correct):
        outcomes = []
        for x in rules:
            outcomes.append(test_rule(x, arrays[array_idx]))

            if outcomes[-1] == False:
                # Correctly position just 1 bad value
                # The next passes will take care of the rest
                a = np.max(np.where(arrays[array_idx] == x[0]))
                b = np.min(np.where(arrays[array_idx] == x[1]))

                el = arrays[array_idx][a]
                arrays[array_idx] = np.delete(arrays[array_idx], a)
                arrays[array_idx] = np.insert(arrays[array_idx], b, el)

        if all(outcomes):
            print(f'Succesfully sorted {arrays[array_idx]}')
            correct = True   
    array_idx = array_idx + 1
```

```python
# Sum up middle values for original properly sorted arrays
mid_sum = 0
for idx, x in enumerate(arrays):
    if all(np.equal(x, orig[idx])):
        mid_sum = mid_sum + x[(len(x)//2)]
print(mid_sum)
```
